```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Contoh data untuk dua survei (Survey 1 dan Survey 2)
# Data ini berisi 5 pertanyaan Likert untuk 4 variabel
data_survey_1 = {
    'Var1': [1, 2, 3, 4, 5, 3, 2, 4, 5, 4],
    'Var2': [5, 4, 3, 2, 1, 3, 2, 4, 5, 3],
    'Var3': [1, 1, 3, 4, 5, 2, 2, 3, 4, 3],
    'Var4': [4, 4, 4, 3, 2, 5, 4, 3, 2, 1],
}

data_survey_2 = {
    'Var1': [3, 2, 5, 4, 1, 4, 5, 3, 2, 1],
    'Var2': [2, 4, 5, 1, 3, 4, 2, 1, 4, 3],
    'Var3': [4, 3, 1, 2, 5, 4, 3, 5, 2, 1],
    'Var4': [5, 4, 3, 1, 2, 5, 3, 4, 1, 2],
}

# Membuat DataFrame
df_survey_1 = pd.DataFrame(data_survey_1)
df_survey_2 = pd.DataFrame(data_survey_2)

# Menggabungkan kedua survei untuk analisis lebih lanjut
df_all = pd.concat([df_survey_1, df_survey_2], keys=['Survey_1', 'Survey_2'])

# Deskriptif Statistics
print("Descriptive Statistics Survey 1:")
print(df_survey_1.describe())
print("\nDescriptive Statistics Survey 2:")
print(df_survey_2.describe())

# Visualisasi
# Membuat boxplot untuk melihat distribusi masing-masing variabel
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_survey_1, palette='coolwarm')
plt.title('Boxplot Survey 1 - Likert Scale Responses')
plt.ylabel('Likert Scale')
plt.xlabel('Variables')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_survey_2, palette='coolwarm')
plt.title('Boxplot Survey 2 - Likert Scale Responses')
plt.ylabel('Likert Scale')
plt.xlabel('Variables')
plt.show()

mean_survey_1 = df_survey_1.mean()
mean_survey_2 = df_survey_2.mean()

print("\nMean for Survey 1:")
print(mean_survey_1)
print("\nMean for Survey 2:")
print(mean_survey_2)

mean_survey_1.plot(kind='bar', title='Average Score per Variable for Survey 1')
plt.ylabel('Mean Likert Score')
plt.show()

mean_survey_2.plot(kind='bar', title='Average Score per Variable for Survey 2')
plt.ylabel('Mean Likert Score')
plt.show()
```
